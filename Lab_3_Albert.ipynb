- Построить график произвольной квадратичной функции

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-10, 10, 400)
y = 6 * x**2 - 3 * x + 5  

plt.plot(x, y)
plt.title('Квадратичная функция: y = 6x^2 - 3x + 5')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True)
plt.show()
```

-  Построить график произвольной гармонической функции

```python

x = np.linspace(0, 10, 400)
y = np.sin(2 * np.pi * x)  

plt.plot(x, y)
plt.title('Гармоническая функция: y = sin(2πx)')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True)
plt.show()
```

- Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

```python
x = np.linspace(-10, 10, 400)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))


ax1.plot(x, 3 * x**2 + 2 * x - 5)
ax1.set_title('Квадратичная функция: y = 3x^2 + 2x - 5')
ax1.set_xlabel('Ось X')
ax1.set_ylabel('Ось Y')
ax1.grid(True)


ax2.plot(x, np.sin(2 * np.pi * x))
ax2.set_title('Гармоническая функция: y = sin(2πx)')
ax2.set_xlabel('Ось X')
ax2.set_ylabel('Ось Y')
ax2.grid(True)


plt.tight_layout()
plt.show()
```

-  Построить графики п.2 и 3 в одной координатной плоскости

```python
plt.plot(x, 2 * x**2 - 3 * x + 1, label='Квадратичная функция')
plt.plot(x, np.sin(2 * np.pi * x), label='Гармоническая функция')
plt.title('Квадратичная и Гармоническая функции')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.legend()
plt.grid(True)
plt.show()
```

- Построить график функции, заданной параметрически

```python

t = np.linspace(0, 4 * np.pi, 400)
x = np.cos(3 * t)
y = np.sin(2 * t)

plt.plot(x, y)
plt.title('Параметрическая функция: x = cos(3t), y = sin(2t)')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True)
plt.show()
```

- Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму

```python
years = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]


plt.bar(years, values, color='skyblue')
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум, сом')
plt.xticks(rotation=45)  
plt.grid(axis='y')  
plt.tight_layout()  

plt.show()
```

- Построить круговой график 

```python

years = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]


plt.pie(values, labels=years, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')
plt.axis('equal')  

plt.show()
```

- Построить график произвольной каркасной поверхности

```python
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.cos(x) * np.sin(y)  #


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(x, y, z, color='green')


ax.set_title('Каркасная поверхность: z = cos(x) * sin(y)')
ax.set_xlabel('Ось X')
ax.set_ylabel('Ось Y')
ax.set_zlabel('Ось Z')

plt.show()
```
